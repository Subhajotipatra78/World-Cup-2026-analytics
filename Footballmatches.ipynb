```python
import numpy as np
import pandas as pd
df = pd.read_csv(r'D:\Projects\FIFA WORLD CUP Project\dataset\results.csv')
df.head()
```

```python
print(f"\nTotal rows and columns: {df.shape}")
```

```python
df.tail
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'] <= '2026-05-09']
df.tail()
```

```python
qualified_teams = [
    'Canada', 'Mexico', 'United States', 'Japan', 'New Zealand', 
    'Iran', 'Argentina', 'Uzbekistan', 'Jordan', 'South Korea', 
    'Australia', 'Brazil', 'Ecuador', 'Paraguay', 'Uruguay', 
    'Colombia', 'Morocco', 'Tunisia', 'Egypt', 'Algeria', 
    'Ghana', 'Cape Verde', 'Qatar', 'Saudi Arabia', 'Senegal', 
    'South Africa', 'Ivory Coast', 'England', 'France', 'Croatia', 
    'Portugal', 'Norway', 'Germany', 'Netherlands', 'Switzerland', 
    'Scotland', 'Spain', 'Austria', 'Belgium', 'Panama', 
    'Curaçao', 'Haiti', 'Bosnia and Herzegovina', 'Sweden', 'Turkey', 
    'Czech Republic', 'DR Congo', 'Iraq'
]
df = df[df['home_team'].isin(qualified_teams) | df['away_team'].isin(qualified_teams)]
df.shape
```

```python
df.tail()
```

```python
df = df[df['tournament'] != 'Friendly']
df.shape
```

```python
df = df.sort_values('date', ascending=False)
recent_matches_list = []
for team in qualified_teams:
    team_matches = df[(df['home_team'] == team) | (df['away_team'] == team)].head(20)
    recent_matches_list.append(team_matches)
df_last_20 = pd.concat(recent_matches_list)
df_last_20 = df_last_20.drop_duplicates()
df_last_20 = df_last_20.sort_values('date', ascending=False)
print(f"New DataFrame size: {df_last_20.shape}")
df_last_20.head()
```

```python
df = df_last_20
conditions = [df['home_score'] > df['away_score'],df['away_score'] > df['home_score']]
choices = [df['home_team'],df['away_team']]
df['Winner'] = np.select(conditions,choices,default='Draw')
df = df.reset_index(drop=True)
df.insert(0,'Serial_No',range(1,len(df)+1))
df.head()
```

```python
df['home_score'] = df['home_score'].fillna(0)
df['away_score'] = df['away_score'].fillna(0)
df['home_score'] = df['home_score'].astype(int)
df['away_score'] = df['away_score'].astype(int)
df['city'] = df['city'].str.title()
df['country'].str.title()
df.isnull().sum()
```
